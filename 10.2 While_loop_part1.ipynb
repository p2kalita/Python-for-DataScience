#### A while loop is used when you want to perform a task indefinitely, until a particular condition is met. It’s a condition-controlled loop.

The while loop enables the Python program to repeat a set of operations while a particular condition is true. When the condition becomes false, execution comes out of the loop immediately, and the first statement after the while loop is executed.

1. Indefinite Iteration: An unknown number of iterations. Ask the user to guess the lucky number. You don’t know how many attempts the user will need to guess correctly. It can be 1, 20, or maybe indefinite. In such cases, use a while loop.


2. Finite Iterations: Fixed number of iterations. Print the multiplication table of 2. In this case, you know how many iterations you need. Here you need 10 iterations. In such a case use for loop.

-------------------------------------------------------------------------------------------------

#### 1. Syntax for while loop

1. The while statement checks the condition. The condition must return a boolean value. Either True or False.

2. Next, If the condition evaluates to true, the while statement executes the statements present inside its block.

3. The while statement continues checking the condition in each iteration and keeps executing its block until the condition becomes false.

#### 2. When do we use a while loop, and why do we use it.

1. Automate and repeat tasks.: As we know, while loops execute blocks of code over and over again until the condition is met it allows us to automate and repeat tasks in an efficient manner.


2. Indefinite Iteration: The while loop will run as often as necessary to complete a particular task. When the user doesn’t know the number of iterations before execution, while loop is used instead of a for loop


3. Reduce complexity: while loop is easy to write. using the loop, we don’t need to write the statements again and again. Instead, we can write statements we wanted to execute again and again inside the body of the loop thus, reducing the complexity of the code


4. Infinite loop: If the code inside the while loop doesn’t modify the variables being tested in the loop condition, the loop will run forever.

#### Example 1. While loop less than 10

```python
var = 1 #initial value for variable
while var<10: #Condition declaration 
    print(var) #print staement for var
    var = var+1 #one will be added until value become less than 10
```

```python
number = int(input("Enter number for Multiplication table: "))
iterate = 1
while iterate<=10:
    print(number, "x", iterate, "=", number*iterate )
    iterate = iterate + 1
```

#### Example 10. Write a while loop to display only alphabets from a string.

In this example, we will print only letters from a string by skipping all digits and special symbols

```python
name = 'Harry123546world'

size = len(name)
i = -1
while i < size - 1:
    i = i + 1
    if not name[i].isalpha():
        continue
    print(name[i], end=' ')
```
